## How often are flights delay?

### Flights delayed which are departing from LAX.

```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


```

```python
# Open and read files
flight_delay_origin = pd.read_csv("../data/clean/LAX_Origin_Flight_Delays_2016.csv")
```

```python
# retrive and conver in to dataframe
flight_delay_origin = pd.DataFrame(flight_delay_origin)
flight_delay_origin.head(10)
```

```python
# Total departure delayed by carriers for each months
table_depdelay = pd.pivot_table(flight_delay_origin, values='DepDel15', index='Month', columns='CarrierName', aggfunc=np.sum)   
table_depdelay
```

```python
# plotting Total departure delayed by carriers for each months
table_depdelay.plot(kind='bar',figsize=(20,12), width=0.7, align='center')
plt.xlabel('Months',fontsize='18')
plt.ylabel('Departur Delays',fontsize='18')
plt.title('Departure carrier delays by Months',fontsize='18')
plt.xticks(np.arange(len(table_depdelay.index)),('Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'))
plt.legend(frameon=True,shadow= True,edgecolor='black',fontsize = 'medium', title='Media Sources')             

#plt.savefig('OriginFlights_Departure carrier delays by months')

plt.show()
```

```python
# Total Arrival delayed by carrier for each month
table_arrdelay = pd.pivot_table(flight_delay_origin, values='ArrDel15', columns='CarrierName', index='Month', aggfunc=np.sum)   
table_arrdelay
```

```python
# Plotting Total Arrival delayed by carrier for each month
table_arrdelay.plot(kind='bar',figsize=(20,12), width=0.7)
plt.legend(frameon=True,shadow= True,edgecolor='black',fontsize = 'medium', title='Media Sources', bbox_to_anchor=(1, 1))
plt.xlabel('Months',fontsize='18')
plt.ylabel('Departur Delays',fontsize='18')
plt.title('Arrival carrier delays by Months',fontsize='18')
plt.xticks(np.arange(len(table_arrdelay.index)),('Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'))
#plt.savefig('OriginFlights_Arrival carrier delays by months')
plt.show()
```

```python
td = flight_delay_origin['CarrierName'].value_counts()
td = pd.DataFrame(td)
td = td.rename(columns={'CarrierName':'Total_Flights'})
td = td.sort_index()
td = td.reset_index()
td = td.rename(columns={'index':'CarrierName'})


dd = table_depdelay.sum()
dd = pd.DataFrame(dd)
dd = dd.rename(columns={0:'Departure_Delayed_Flights'})
dd = dd.reset_index()
dd = pd.DataFrame(dd)



ad = table_arrdelay.sum()
ad = pd.DataFrame(ad)
ad = ad.rename(columns={0:'Arrival_Delayed_Flights'})
ad = ad.reset_index()
ad = pd.DataFrame(ad)
```

```python
#Total arrival and departure delays vs. Total flights per carrier.
del_table_total = pd.merge(dd,td)
del_table_total = pd.merge(del_table_total,ad)
del_table_total
```

```python
#Total departure delays in 2016
del_table_total.Departure_Delayed_Flights.sum()
```

```python
#Total arrival delays in 2016
del_table_total.Arrival_Delayed_Flights.sum()
```

```python
#Total flights in 2016
del_table_total.Total_Flights.sum()
```

```python
# plotting total flights, total arrival delays and total departure delays in 2016

del_table_total.plot(kind='bar',figsize=(20,12), width=0.7, align='center')
plt.xticks(np.arange(len(del_table_total.CarrierName)),('Alaska Airlines Inc.', 'American Airlines Inc.',
       'Delta Air Lines Inc.', 'Frontier Airlines Inc.',
       'Hawaiian Airlines Inc.', 'JetBlue Airways', 'SkyWest Airlines Inc.',
       'Southwest Airlines Co.', 'Spirit Air Lines', 'United Air Lines Inc.',
       'Virgin America'))
plt.xlabel('Carriers', fontsize=15)
plt.ylabel('Delays', fontsize=15)
plt.title('Total origin flights delays in 2016',fontsize=20)
plt.savefig("Origin flights delayed in 2016")
plt.show()
```
